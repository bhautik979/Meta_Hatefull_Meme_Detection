# Download Multiple Kaggle Notebook Outputs
This notebook shows how to download outputs from several Kaggle notebooks in a loop using PowerShell.

```python
$notebooks = @("08-training", "09-auroc-optimization", "09b-auroc-unfreeze-distill", "09c-vitl14-upgrade")
foreach ($nb in $notebooks) {
    kaggle kernels output "muddybuddy/$nb" -p "./kaggle-outputs/$nb"
}
```

```python
# List all downloaded outputs
Get-ChildItem ./kaggle-outputs -Recurse
```
